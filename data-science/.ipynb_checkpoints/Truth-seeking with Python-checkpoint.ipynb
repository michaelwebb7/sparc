```python
##########################
## IMPORT LIBRARIES
##########################

import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns

from __future__ import division

%matplotlib inline

##########################
## IMPORT THE DATA
##########################

reader = pd.io.stata.StataReader('https://github.com/michaelwebb7/sparc/raw/master/data-science/sparc_dataset.dta?raw=true')

# Import the data as dataframe df
df = reader.read()
vf = pd.read_stata('https://github.com/michaelwebb7/sparc/raw/master/data-science/sparc_dataset.dta?raw=true',convert_categoricals=False)

# Store Stata variable labels as a dictionary
variable_dict = reader.variable_labels()

# Store Stata value labels in a dictionary
value_dict = reader.value_labels()

##########################
## DEFINE FUNCTIONS
##########################

# Define a variable lookup function
def lookfor(searchFor):
    return [[key, value] for key, value in variable_dict.items() if searchFor in value.lower()]

# Define a variable typer
def vtype(var):
    '''Describe variable type.'''
    return getattr(df, var).dtype

# Define the info function
def info(var):
    '''Information on a variable.'''
    if var in value_dict.keys():
        v = value_dict[var]
    else:
        v = 'Not a category.'
    return [variable_dict[var], getattr(df, var).dtype, v]

# Define the sum
def sum(var,subset=None):
    '''Summarize a variable.
    
    Args:
        var: Variable to be tabulated.
        subset: Optional subset of the data.
    
    Returns:
        A pandas DataFrame containing summary statistics.
    '''
    
    # Subset table if required
    if subset is not None:
        ff = df.query(subset)
    else:
        ff = df
    
    ff = pd.DataFrame(getattr(ff, var))
    
    j = ff.describe(include='all')
    j.columns = [variable_dict[var]]
    return np.round(j.transpose(), decimals=2)

# Define the tab
def tab(var,subset=None):
    '''Tabulate one variable.
    
    Args:
        var: Variable to be tabulated.
        subset: Optional subset of the data.
    
    Returns:
        A pandas DataFrame. First column contains frequencies, second contains percentages.
    '''
    
    # Subset table if required
    if subset is not None:
        ff = df.query(subset)
    else:
        ff = df
    
    # Get variable column (a series) and store as dataframe
    j=pd.DataFrame(getattr(ff, var))
    
    # Make column of frequencies
    j=j.apply(lambda x: x.value_counts(dropna=False)) 
    
    # Make column of percentages
    k=np.round(j.apply(lambda r: 100*r/r.sum()), decimals=2)
    
    # Combine and label columns
    l=pd.concat([j,k],axis=1)
    l.columns = ['Freq.','Percent']
    l.columns.name = variable_dict[var]    
    
    # Return final table
    return l

# Define the crosstab
def crosstab(rowvar,colvar,axis=None,subset=None):
    '''Cross-tabulate two variables.
    
    Args:
        rowvar: First variable to be cross-tabulated. Displayed in rows.
        colvar: Second variable to be cross-tabulated. Displayed in columns.
        axis: Defines table cell contents.
                If empty, returns frequency counts.
                If 'row', returns the relative (%) frequency of each cell within its row.
                If 'col', returns the relative (%) frequency of each cell within its column.
        subset: Optional subset of the data.
    
    Returns:
        A pandas DataFrame.
    '''
    
    # Subset table if required
    if subset is not None:
        ff = df.query(subset)
    else:
        ff = df
    
    if axis is 'cols':
        j = pd.crosstab(getattr(ff, rowvar),getattr(ff, colvar)).apply(lambda r: 100*r/r.sum(), axis=0)
    elif axis is 'rows':
        j = pd.crosstab(getattr(ff, rowvar),getattr(ff, colvar)).apply(lambda r: 100*r/r.sum(), axis=1)
    else:
        j = pd.crosstab(getattr(ff, rowvar),getattr(ff, colvar))
    j.index.name = variable_dict[rowvar] + ':'
    j.columns.name = variable_dict[colvar] + '>>'
    return np.round(j, decimals=2)

# Define the histogram plotter
def hist(var,subset=None):
    '''Plot an histogram.
    
    Args:
        var: Variable whose histogram is to be plotted.
    
    Returns:
        An histogram.
    '''
    
    # Subset table if required
    if subset is not None:
        qf = vf.query(subset)
        pf = df.ix[list(qf.index)]
    else:
        pf = df
        
    fig = plt.figure()
    getattr(pf, var).plot.hist()
    fig.suptitle(variable_dict[var], fontsize=20)

# Define the plot function
def plot(var):
    '''Plot a bar chart of a variable.
    
    Args:
        var: Variable whose bar chart is to be plotted.
    
    Returns:
        A bar chart.
    '''
    fig = plt.figure()
    getattr(df, var).value_counts().plot.bar()
    plt.xticks(rotation=20)    
    fig.suptitle(variable_dict[var], fontsize=20)
    
def plotordcat(x,y,split=None,subset=None):

    # Subset table if required
    if subset is not None:
        qf = vf.query(subset)
        pf = df.ix[list(qf.index)]
        getattr(pf, x).cat.remove_unused_categories(inplace=True)
    else:
        pf = df

    #fig = plt.figure()
    
    if split is not None:
        ax = sns.pointplot(x=x, y=y, hue=split, data=pf);
    else:
        ax = sns.pointplot(x=x, y=y, data=pf);

    ax.set(xlabel=variable_dict[x],title=variable_dict[y])
    plt.xticks(rotation=20)
    plt.show()
    
def plotcat(x,y,split=None,subset=None):

    # Subset table if required
    if subset is not None:
        qf = vf.query(subset)
        pf = df.ix[list(qf.index)]
        getattr(pf, x).cat.remove_unused_categories(inplace=True)
    else:
        pf = df

    #fig = plt.figure()
    
    if split is not None:
        ax = sns.factorplot(x=x, y=y, hue=split, data=pf, kind='bar');
    else:
        ax = sns.factorplot(x=x, y=y, data=pf, kind='bar');

    ax.set(xlabel=variable_dict[x],title=variable_dict[y])
    plt.xticks(rotation=20)
    plt.show()
    
def values(var):
    '''Give the value labels of a variable.
    
    Args:
        var: Variable.
    
    Returns:
        Dictionary of value labels.
    '''
    return value_dict[var]
```

## The dataset: Add Health
> The National Longitudinal Study of Adolescent to Adult Health (**Add Health**) is a **longitudinal** study of a **nationally representative sample** of adolescents in grades 7-12 in the United States during the 1994-95 school year. The Add Health cohort has been followed into young adulthood with four in-home interviews, the most recent in 2008, when the sample was aged 24-32. Add Health combines longitudinal survey data on respondents’ social, economic, psychological and physical well-being with contextual data on the family, neighborhood, community, school, friendships, peer groups, and romantic relationships, providing unique opportunities to study how social environments and behaviors in adolescence are linked to health and achievement outcomes in young adulthood.

## Topics covered
### Wave 2 (14-18 years old)
Lifestyle; health, exercise, and sleep; food/drink consumed yesterday; meals in last week; school grades, behavior, and attitudes; parents’ smoking; relationship with parents; feelings about parents; personality; male friends; female friends; romantic relationships; feelings about adults; religion; college and work; predictions for the future; interviewer’s assessment of respondent.
### Wave 4 (27-31 years old)
Relationship with parents; friendships; health; housing; education; job; job characteristics; income and finance; religion; children; intelligence tests; sexuality; lifestyle; personality; height and weight; interviewer’s assessment of respondent.

## Crosstabulating two variables

What if we want to examine the *joint* distribution of two variables? We have a command for that too!

First, let's see what variables there are about religion:

```python
lookfor('religion')
```

Wouldn't it be interesting to see how people's religion in high school is related to their religion in later life?

Do you expect it to stay the same or to change?

Let's see!

```python
crosstab('religion_highschool','religion_adult')
```

The table above gives us frequencies (counts) for each cell in the table.

This is interesting, but it would be nicer if we could look at *relative* frequencies. We can specify a third argument for the *crosstab* function, that calculates relative frequencies either by row or column. Let's use row here.

```python
crosstab('religion_highschool','religion_adult','rows')
```

## Last couple of commands

Where a variable's categories have a natural 'order', we can use the **plotordcat** command to plot how the values of some *other* variable change with the first variable.

For example, we can plot how much sleep people get (y-axis) against their most recent math grade (x-axis):

```python
plotordcat('H2ED8','H2GH44')
```

We'd rather not include those various 'miscellaneous' categories. So we can subset!

First, use the **info** command to work out which numbers correspond to the categories we want to keep:

```python
values('H2ED8')
```

OK, so we want to keep values that are < 5. While we're at it, let's also only keep 'possible' values of sleeping hours (< 24):

```python
plotordcat('H2ED8','H2GH44',subset='H2ED8<5 & H2GH44<10')
```

** Can you come up with an explanation for the pattern we observe? **

Finally, we can also 'split' the data along some third dimension. For example:

```python
plotordcat('H2ED8','H2GH44',split='female',subset='H2ED8<5 & H2GH44<10')
```

So **boys get more sleep**, it seems!

Lastly, we have a very similar function, **plotcat**, that takes exactly the same arguments, except that the first argument (that goes on the x-axis) is not an ordered category. Let's use it to look how income in later life (wave 4) is a function of the number of close friends you have as a teenager (wave 2):

```python
plotcat('H4WS4','personal_income',subset='H4WS4<95')
info('H4WS4')
```

Wow, that's quite a pattern! Let's check whether it's true for both men and women:

```python
plotcat('H4WS4','personal_income',split='female',subset='H4WS4<95')
```

Yes - the pattern is very similar. But women earn less at every level! Why would this be?

Note that for **plotcat** and **plotordcat**, the second variable (that goes on the y-axis) should be numeric, not categorical. Some numeric variables include:
